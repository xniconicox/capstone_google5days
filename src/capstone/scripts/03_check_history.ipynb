# check eval result



```python
import json
import os 
def get_latest_eval_path(history_dir="./eval/.adk/eval_history"):
    # ディレクトリ内のファイル一覧を取得
    files = [
        os.path.join(history_dir, f)
        for f in os.listdir(history_dir)
        if os.path.isfile(os.path.join(history_dir, f))
    ]

    if not files:
        return None

    # 最終更新時刻が最新のファイルを選択
    latest_file = max(files, key=os.path.getmtime)
    return latest_file

file_dir = "./eval/.adk/eval_history"
latest_path = get_latest_eval_path(file_dir)
output_path = "./eval/tmp/evalset_result_pretty.json"

with open(latest_path, "r") as f:
    raw = f.read()     # ① Raw text (a string that includes a JSON string)
    outer = json.loads(raw)  # ② outer should be *a string*

# If outer is a str, apply json.loads again
if isinstance(outer, str):
    data = json.loads(outer)  # ③ This will become a dict
else:
    data = outer

with open(output_path, "w") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("written:", output_path)
```
